```python
import re
```

```python
#data cleaning
# Remove special characters and punctuation using regex
mydata=[] #empty list
for i in data['Message']:
  text = re.sub(r',','',i)
  cleaned_text = re.sub(r'[^\w\s]', '', text)  # Removes everything except word characters and spaces
  cleaned_text = re.sub(r'\s+', ' ', cleaned_text)  # Replaces multiple spaces with a single space
  cleaned_data = cleaned_text.strip()  # Removes leading/trailing whitespace
  mydata.append(cleaned_data)
```

```python
mydata
```

```python
from nltk.tokenize import word_tokenize
```

```python
#data tokenization
tokens=[word_tokenize(x) for x in mydata]
```

```python
tokens
```

```python
#to remove stopwords
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stop = stopwords.words('english')
```

```python
tokens[0]
```

```python
cleaned_tokens=[]
for x in tokens:
  cl = [i for i in x if i not in stop]
  cleaned_tokens.append(cl)
```

```python
from nltk.tokenize import sent_tokenize
```

```python
sent_tokens=[sent_tokenize(i)  for i in data["Message"]]
sent_tokens[6]
```

```python
# Removing stopwords from tokenized words
# Stopwords are commonly used words like "is", "the", "and", etc., which are removed to reduce noise
stop = set(stopwords.words('english'))  # Fetching the list of English stopwords
stop_token = []  # List to store stopword-removed tokens
for k in range(len(data['Message'])):  # Loop through the tokenized text
    p = [i for i in tokens[k] if i not in stop]  # Filter out tokens that are in the stopword list
    stop_token.append(p)  # Append the filtered tokens to the list
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
ps.stem("studied")
```

```python
stemedata =[]
for message in stop_token:
  st = [ps.stem(word) for word in message]
  stemedata.append(st)
```

```python
stemedata
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
lemmatizer=WordNetLemmatizer()
```

```python
lemmatizer.lemmatize("studying","v")
```

```python
from nltk.corpus import wordnet

def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN  # Default to NOUN if POS tag is unknown
```

```python
lem_data = []
for x in range(len(pos_token)):
  lem = [lemmatizer.lemmatize(pos_token[x][i][0],get_wordnet_pos(pos_token[x][i][1])) for i in range(len(pos_token[x]))]
  lem_data.append(lem)
```
